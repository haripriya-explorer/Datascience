```python
#grouped barchart
import matplotlib.pyplot as plt
import numpy as np
categories=["zebra","horse","elephant"]
values1=[15,24,30]
values2=[20,18,25]
bar_width=0.35
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1 + bar_width
plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='purple')
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='orange')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('grouped bar chart')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=['transport','income tax','house']
v1=[15,24,30]
v2=[20,18,25]
plt.bar(categories,v1,label='first half',color="red")
plt.bar(categories,v2,bottom=v1,label='second half',color="green")
plt.xlabel('categories')
plt.ylabel('values')
plt.title('stacked bar chart')
plt.legend()
plt.show()
```

```python
data={'Name':['Alice','bob','Ave'],
      'Age':[25,30,35],
      'city':['New york','Los Angeles','Chicago']}
df=pd.DataFrame(data)
print(df)
print('------------------')
#selecting columns')
print(df.Name)
print(df['Name'])
print('-------------------')
print(df[['Name','city']])
print('----------------')
#select a single row by label
print("1.\n",df.loc[0])
print("2.\n",df.iloc[0])
print('--------------')
#conditional selection
print('conditional')
print(df[df['Age']>25])
print(df.loc[(df.Name=='bob')&(df.city=='Los Angeles')])
```

```python
import pandas as pd
data={
    'name':['Alice','Bob','charlie','Alice','Bob'],
    'age':[10,20,30,10,20],
    'city':['Newyork','francisco','los angeles','Newyork','francisco'],
    'salary':[70000,80000,120000,70000,80000]
}
df=pd.DataFrame(data)
grouped=df.groupby('name')
print(grouped.groups)
print("******")
#calculate the sum of salaries for each group
salary_sum= grouped['salary'].sum()
print(salary_sum)
print("*******")
age_mean=grouped['age'].mean()
print(age_mean)
print("------------------")
#Grouping by multiple columns
#group by 'name' and 'city'
grouped_multiple=df.groupby(['name','city'])
#calculate the sum of salaries for each group
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```

```python
#pandas merging,joining,conacatenation of the dataframes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[4,5,6,7])
print(df,"\n\n",df1)
print("------------------")
#concatenation
frames=[df,df1]
result=pd.concat(frames)
print(result)
```

```python
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res1=pd.concat([df,df1])
print(res1)
res2=pd.concat([df,df1],axis=1,join='inner')
print(res2)
```

```python
#joining of 2 datframes by ignoring the indexes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
data2={'Name':['Abhi','sai','cool','jani'],
       'Age':[21,22,43,56],
       'Address':['hyd','vzg','nellore','ongole'],
       'Qualification':['B.Tech','B.A','B.Com','B.hons']}
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n",df1)
res=pd.concat([df,df1],ignore_index=True)
print(res)
```

```python
#concatenate a mix of series and dataframes
import pandas as pd
data1={'Name':['jai','princi','gaurav','anuj'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','karnataka'],
       'Qualification':['msc','ma','btech','phd']}
#convert the dictionary into datframe
df=pd.DataFrame(data1,index=[0,1,2,3])
#creating the series
s1=pd.Series([1000,2000,3000,4000],name='Salary')
#concatenate the series and dataframe
result=pd.concat([df,s1],axis=1)
print(result)
```

```python
#Merging the dataframes
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'key':['k0','k1','k2','k3'],
       'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
       'Qualification':['Btech','B.A','Bcom','B.hons']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
res=pd.merge(df,df1,on='key')
print(res)
```

```python
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'key1':['k0','k1','k0','k1'],
       'Name':['Jai','Princi','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'key':['k0','k1','k2','k3'],
       'key1':['k0','k0','k0','k0'],
       'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
       'Qualification':['Btech','B.A','Bcom','B.hons']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on=['key','key1'])
print(res)
```

```python
import pandas as pd
data1={'name1':['cool','jani','raju','anil'],
       'email1':['cooluu@gmail.com','jani@gmail.com','raju12@gmail.com','anil@gmail.com'],
       'age1':[12,34,21,34],
       'city':['ongole','vijayawada','kadapa','kurnool'],
       'address':['bandarroad','nellore bustand','ramnagar','bhagyanagar']}
data2={'name2':['hari','sai','priya','anjali'],
       'email2':['hari@gmail.com','sai@gmail.com','priya@gmail.com','anjali@gmail.com'],
       'age2':[21,34,54,32],
       'city':['ongole','nellore','kavali','kurnool'],
       'address':['mm road','kurnool road','trunk road','dharavarithota']}

df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge (df,df1,on='city',how='left',indicator="True",suffixes=('_left','_right'))
print(res)
```
